```python
import numpy as np 
import matplotlib.pyplot as p

M = 100
NN = np.arange(1000,1e6+1,1000, int)
med = np.zeros(NN.shape[0])
var = np.zeros(NN.shape[0])
ppp = np.zeros(NN.shape[0])

for i in range(NN.shape[0]):
    
    arr = np.random.randint(1,M+1,NN[i])
    med[i] = arr.mean()
    var[i] = (arr.std())**2
    ppp[i] = len(arr[np.where(arr<=60)])/len(arr)
    

```

```python
print(med[0])
print(med[9])
print(med[-1])
print('\n')
print(99/2)
print('\n')

print(var[0])
print(var[9])
print(var[-1])
print('\n')
print((100**2-1)/12)
print('\n')

print(ppp[0])
print(ppp[9])
print(ppp[-1])
print('\n')
print(0.6)
```

```python
p.figure(figsize = (6,5))
p.plot(NN,med,'.',label = 'Numerical Calculation')
p.plot(NN,(M+1)/2*np.ones(NN.shape[0]),'--',c='red',label = 'Theoretical Value $<x> = $' + str((M+1)/2))
p.title('Mean Value')
p.xlabel('N')
p.ylabel(r'$<x>$')
p.legend()
p.savefig('mean.eps',format = 'eps')
p.show()

p.figure(figsize = (6,5))
p.plot(NN,var,'.', label='Numerical Calculation')
p.title('Variance')
p.plot(NN,(M**2-1)/12*np.ones(NN.shape[0]),'--',c='red',label = 'Theoretical Value $<x^2>-<x>^2 = $' + str((M**2-1)/12))
p.xlabel('N')
p.ylabel(r'$<x^2>-<x>^2$')
p.legend()
p.savefig('var.eps',format = 'eps')
p.show()

p.figure(figsize = (6,5))
p.plot(NN,ppp,'.', label='Numerical Calculation')
p.title('p value')
p.plot(NN,(0.6)*np.ones(NN.shape[0]),'--',c='red',label = 'Theoretical Value $p = $' + str((0.6)))
p.xlabel('N')
p.ylabel(r'$p$')
p.legend()
p.savefig('ppp.eps',format='eps')
p.show()

```

```python
NN1 = np.arange(1000,1e6+1,1000, int)

mean1 = np.zeros(NN1.shape[0])
mean2 = np.zeros(NN1.shape[0])
mean3 = np.zeros(NN1.shape[0])

for i in range(NN1.shape[0]):
    x1 = np.random.rand(NN1[i])
    x2 = np.random.rand(NN1[i])
    z = x1*x2
    mean1[i] = x1.mean()
    mean2[i] = x2.mean()
    mean3[i] = z.mean()
```

```python
print(mean1[0])
print(mean1[9])
print(mean1[-1])
print('\n')

print(mean2[0])
print(mean2[9])
print(mean2[-1])
print('\n')

print(mean3[0])
print(mean3[9])
print(mean3[-1])
print('\n')
```

```python
p.figure(figsize = (6,5))
p.plot(NN1,(mean1*mean2-mean3)/mean3*100,'.',label = 'Numerical Calculation')
p.plot(NN1,0*np.ones(NN1.shape[0]),'--',c='red',label = r'$\varepsilon=0$')
p.title('Error')
p.xlabel('N')
p.ylabel(r'$\varepsilon$')
p.legend()
p.savefig('mean1.eps',format='eps')
p.show()



```

```python
import numpy as np 
import matplotlib.pyplot as p

for N in [100,1000,1000000]:
    probs = np.random.rand(N)
    print('mean '+str(probs.mean()))
    print('err mean '+str(np.abs( probs.mean()/0.5-1 )*100 ) )
    print('\n')
    print('var '+str((probs.std())**2))
    print('err mean '+str(np.abs( probs.std()**2*12-1 )*100 ) )
    dy = 0.005
    binada = np.arange(0,1+dy,dy)
    bins = p.hist(probs, bins = binada, density = True,label=r'Numerical $p(x)$')
    p.plot(np.linspace(0,1,1000),np.ones(1000),'--',color='red',label = r'Theoretical $p(x)$')
    p.title('N = ' + str(N))
    p.xlabel('x')
    p.ylabel('$p(x)$')
    p.legend()
    p.savefig('N'+str(N)+'.jpg')
    p.show()
```

```python
N = int(1e6)
arrx = np.linspace(0,1,1000)

for  n in [10,100,1000]:
    probg = np.random.rand(N,n)
    probY = probg.mean(axis = 1)
    meanx=probg.mean()
    meanY=probY.mean()
    varx=probg.std()**2
    var= probY.std()**2
    limvar=varx/n
    error= np.abs(var/limvar-1)
    error1 = np.abs(meanx/meanY-1)
    print('MEAN')
    print('\n')
    print('mean x '+str(meanx))
    print('mean Y '+str(meanY))
    print('error '+str(error1))
    print('\n')
    
    print('VARIANCE')
    print('\n')
    print('var '+str(var))
    print('lim var '+str(limvar))
    print('error '+str(error))
    print('\n')
    print('#----------------------#')
    dy = 0.005
    binada = np.arange(0,1+dy,dy)
    bins = p.hist(probY, bins = binada, density = True,label='n = ' + str(n))
    p.plot(arrx,np.exp(-0.5*(arrx-meanY)**2/var)/np.sqrt(2*np.pi*var),'--',label = r'Gaussian $\sigma ^2 = $ '+"{:.0e}".format(var))
p.title(r'Histogram of Y ($N=10^6$)')
p.xlabel('Y')
p.ylabel('$p(Y)$')
p.legend()
p.savefig('yvar.eps',format='eps')
p.show()
```

```python
import numpy as np
import matplotlib.pyplot as p
import scipy.special as sp

M = 9
N = 21 
#binada2 = np.arange(0.75,25+0.5,0.5)
binada2 = np.arange(-0.5,21+0.6,1)
pp = 1/M

mean = N*pp

n=np.arange(0,N+1)

for K in [10**3,10**4,10**6]:
    probarr = np.random.randint(1,M+1,(K,21))
    mc = np.where(probarr==3,probarr*0+1,probarr*0)
    counts = mc.sum(axis=1)
    
#     bins = np.histogram(counts, bins = binada2)
#     p.plot(n,bins[0]/bins[0].sum(),'.',label = 'Numeric Values')
    
    bins = p.hist(counts, bins = binada2,density = True,label = 'Histogram')
    
    p.plot(n,sp.binom(N,n)*pp**n*(1-pp)**(N-n),label = 'Binomial')
    p.plot(n,(N*pp)**n/sp.factorial(n)*np.exp(-N*pp),label = 'Poisson')
    p.plot(n,np.exp(-0.5*(n-mean)**2/mean)/np.sqrt(2*np.pi*mean),label = 'Gaussian')
    
    p.title('K = '+str(K))
    p.xlabel('n')
    p.ylabel(r'P(n)')
    p.xlim(-1,15)
    
    p.legend()
    p.savefig('k'+str(K)+'.eps',format = 'eps')
    p.show()
    
```
